```python
def func(bc):
    l=list(map(str,bc.split(';')))
    return l
```

```python
def listreturn(l):
    i=0
    p=[[l[2][0],l[2][1]]]
    fl=0
    while i<len(l):
        if len(l[i])==8:
            if l[i][0]==p[fl][0]:
                p[fl].append(float(l[i][4]))
            else:
                k=[l[i][0],l[i][1],float(l[i][4])]
                p.append(k)
                fl+=1
        i+=1
    return p
```

```python
dfl = listreturn(li)
```

```python
dfl1 = ['Scheme Id','Scheme Name']
i=2
s=li[2][0]
while i<len(li):
    if s!=li[i][0]:
        break
    dfl1.append(li[i][7])
    i+=1
```
