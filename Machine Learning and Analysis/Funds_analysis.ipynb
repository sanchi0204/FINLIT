```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
high_risk = pd.read_csv('High_risk.csv',index_col='Scheme Name')
very_high_risk = pd.read_csv('very_high_risk.csv',index_col='Scheme Name')
low_mod_risk = pd.read_csv('moderate_low_risk.csv',index_col='Scheme Name')
```

<h2>Transposing the data in order to apply percentage change function</h2>

```python
new_df = high_risk.drop('Scheme Id',axis=1,inplace=False).T
n_df = very_high_risk.drop('Scheme Id',axis=1,inplace=False).T
nw_df = low_mod_risk.drop('Scheme Id',axis=1,inplace=False).T
n_df
```

<h2>Detecting percentage change after each sample</h2>

```python
pc = new_df.pct_change()
pc1 = n_df.pct_change()
pc2 = nw_df.pct_change()
```

<h2>Mean of pc --> Average Return Rate over the period of last 7 months</h2>
<h2>std of pc --> signifies how deviated examples are from the mean ie volatility of the scheme (Risk)</h2>

```python
plt.figure(figsize=(10,6),dpi=100)
plt.scatter(pc.mean(),pc.std(),color='#ff8b3d',label='LargeCap Equity')
#for label,x,y in zip(pc.columns,pc.mean(),pc.std()):
#    plt.annotate(label,(x,y),textcoords="offset points",xytext=(0,10),ha='center')
plt.scatter(pc1.mean(),pc1.std(),color='r',label='Small and Midcap Equity')
plt.scatter(pc2.mean(),pc2.std(),color='#47DBCD',label='Debt Funds')
plt.ylabel('Risk')
plt.xlabel('Expected Returns')
plt.title('Risk vs Returns')
plt.legend()
```

<h2>Annualised Return Rate</h2>

```python
bgh = ((1+(bgh/100))**(365/210))-1
bgh1 = ((1+(bgh1/100))**(365/210))-1
bgh2 = ((1+(bgh2/100))**(365/210))-1
```

```python
xcfg = pd.concat([pc.mean()*10000,pc.std()*10000,bgh*100],axis=1)
xcfg
```

<h2>Dividing Growth and IDCW schemes</h2>

```python
xc = pd.concat([pc.mean()*10000,pc.std()*10000,bgh*100],axis=1)
xc.columns=['Expected Returns','Risk','1 yr Return']
xc_growth = pd.DataFrame([xc.iloc[i,:] for i in range(0,len(pc.std())) if xc.index[i].find('Growth')!=-1 or xc.index[i].find('GROWTH')!=-1]) 
xc_idcw = pd.DataFrame([xc.iloc[i,:] for i in range(1,len(pc.std())) if xc.index[i].find('Growth')==-1 and xc.index[i].find('GROWTH')==-1])
xc_growth
```

```python
len(xc)==len(xc_idcw)+len(xc_growth)
```
